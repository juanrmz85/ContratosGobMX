solventes en sobre cerrado, que será abierto públicamente, a fin de asegurar al Estado las mejores

Sesión 1: Estimados de Locación y Variabilidad

Partiendo del proyecto final del modulo 3, teniendo ya nuestro dataset limpio, procederemos a obtener estimados de locación y variabilidad para describir nuestros datos.

*Promedio*

```python
round(contratos_clean['Importe del contrato'].mean(), 2)
```

*Mediana*

```python
contratos_clean['Importe del contrato'].median()
```

*Media Truncada*

```python
from scipy import stats
```

```python
round(stats.trim_mean(contratos_clean['Importe del contrato'], 0.1), 2)
```

*Desviación Estandar*

```python
round(contratos_clean['Importe del contrato'].std(), 2)
```

*Rango*

```python
contratos_clean['Importe del contrato'].max() - contratos_clean['Importe del contrato'].min()
```

*Percentiles*

```python
print(f'Valor mínimo: {contratos_clean["Importe del contrato"].min()}')
print(f'Percentil 10: {contratos_clean["Importe del contrato"].quantile(0.1)}')
print(f'Percentil 25: {contratos_clean["Importe del contrato"].quantile(0.25)}')
print(f'Percentil 50: {contratos_clean["Importe del contrato"].median()}')
print(f'Percentil 75: {contratos_clean["Importe del contrato"].quantile(0.75)}')
print(f'Percentil 90: {contratos_clean["Importe del contrato"].quantile(0.9)}')
print(f'Valor máximo: {contratos_clean["Importe del contrato"].max()}')
```

*Rango intercuartílico*

```python
round(contratos_clean['Importe del contrato'].quantile(0.75) - contratos_clean['Importe del contrato'].quantile(0.25),2)
```

Los boxplots los podemos utilizar para visualizar la distribución de nuestros valores de acuerdo a la media de los mismos.

```python
import seaborn as sns
```

```python
sns.set(style="whitegrid")
sns.boxplot(x=contratos_clean['Importe del contrato'])
```

Para poder visualizar este análisis de una forma mas acotada, delimitamos los valores atípicos con el rango intercuartílico.

```python
iqr = contratos_clean['Importe del contrato'].quantile(0.75) - contratos_clean['Importe del contrato'].quantile(0.25)
filtro_inferior = contratos_clean['Importe del contrato'] > contratos_clean['Importe del contrato'].quantile(0.25) - (iqr * 1.5)
filtro_superior = contratos_clean['Importe del contrato'] < contratos_clean['Importe del contrato'].quantile(0.75) + (iqr * 1.5)

df_filtrado = contratos_clean[filtro_inferior & filtro_superior]
```

```python
sns.boxplot(df_filtrado['Importe del contrato'])
```

Tomando en cuenta la amplitud que tenemos en nuestros datos, podemos dividir nuestro conjunto de datos en segmentos.

```python
importes = contratos_clean['Importe del contrato']
```

```python
segmentos = pd.cut(importes, 10)

contratos_clean['Importe del contrato'].groupby(segmentos).count()
```

Con este método podemos conocer la cantidad de procedimientos que se llevaron a cabo de acuerdo a su tipo de adjudicación.

```python
conteo = contratos_clean['Tipo de procedimiento'].value_counts()
display(conteo)
```

```python
ax = sns.barplot(conteo)
ax.set_title('Conteo por tipo de procedimiento')
ax.set(ylabel='conteo')
ax.set_xticklabels(ax.get_xticklabels(), rotation=85);
```

Hacemos una conversión de los valores para presentarlos en porcentaje.

```python
conteo * 100 / conteo.sum()
```

```python
en_porc = conteo * 100 / conteo.sum()
```

```python
ax = sns.barplot(en_porc)
ax.set_title('Conteo por tipo de procedimiento')
ax.set(ylabel='conteo')
ax.set_xticklabels(ax.get_xticklabels(), rotation=85);
```
