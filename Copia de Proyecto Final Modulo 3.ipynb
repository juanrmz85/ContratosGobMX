**Introducción:**

El artículo 134 de la Constitución Política de los Estados Unidos Mexicanos establece que los recursos económicos de que dispongan la Federación, las entidades federativas,
los Municipios y las demarcaciones territoriales de la Ciudad de México, se administrarán con eficiencia,
eficacia, economía, transparencia y honradez para satisfacer los objetivos a los que estén destinados.

Así mismo establece que las adquisiciones, arrendamientos y enajenaciones de todo tipo de bienes, prestación de servicios de
cualquier naturaleza y la contratación de obra que realicen, se adjudicarán o llevarán a cabo a través de
licitaciones públicas mediante convocatoria pública para que libremente se presenten proposiciones
solventes en sobre cerrado, que será abierto públicamente, a fin de asegurar al Estado las mejores
condiciones disponibles en cuanto a precio, calidad, financiamiento, oportunidad y demás circunstancias
pertinentes.

Conociendo esto, en el siguiente proyecto haremos el análisis de una base de datos del Gobierno de México de los contratos derivados de los procedimientos de contratación reportados en CompraNet por las unidades compradoras en 2020.  

**Importamos Librerias**

Como punto inicial de nuestro cuaderno, importamos librerias y montamos el drive que estaremos utilizando durante el desarrollo del proyecto.

```python
import pandas as pd
import numpy as np
```

**Importamos Datos**

En este punto necesitamos importar los datos con los cuales estareamos trabajando durante el proyecto.

```python
contratos = pd.read_csv('/content/drive/MyDrive/Checkpoint Modulo 3/Contratos2020.csv',index_col=0)

contratos.head()
```

**Limpieza de Datos**

Para poder proceder con la limpieza de nuestros datos, primero hacemos un analisis para saber el tamaño de nuestro DS, así como el tipo de datos con los cuales estaremos trabajando.

```python
contratos.shape
```

```python
contratos.dtypes
```

Utilizando la función info del DF, podremos ver mas información de cada columna.

```python
contratos.info()
```

Después de visualizar el conteo de los valores nulos en nuestra base de datos, haremos un conteo de los valores N/A presentes en nuestra información.

```python
contratos.isna()
```

```python
contratos.isna().sum(axis=0)
```

Para poder trabajar con un DF más reducido y que contenga únicamente los valores de mayor interés para el análisis propuesto, vamos a eliminar todas las columnas que contienen valores N/A.

```python
contratos_clean = contratos.dropna(axis=1, how='any')
```

```python
contratos_clean.isna()
```

```python
contratos_clean.isna().sum(axis=0)
```

Nuestro nuevo DF quedó conformado con un total de 17 columnas y 159,726 filas.

```python
contratos_clean.shape
```

```python
contratos_clean.head()
```

Guardamos el resultado para tener el documento disponible para otras consultas:

```python
contratos_clean.to_csv('/content/drive/MyDrive/Checkpoint Modulo 3/Contratos2020_clean.csv')
```

**Realizamos un análisis exploratorio de los datos**

Ya con nuestro DF limpio podemos hacer ciertas consultas para contestar algunas preguntas.

```python
contratos_clean.groupby('Tipo de procedimiento').size()
```

```python
contratos_clean.groupby('Tipo de procedimiento')['Siglas de la Institución'].value_counts()
```

```python
contratos_clean.groupby(['Tipo de procedimiento', 'Siglas de la Institución'])['Importe del contrato'].sum()
```
